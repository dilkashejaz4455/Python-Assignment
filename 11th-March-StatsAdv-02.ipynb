**Type 1 Error Example:**
In a legal context, if you wrongly convict an innocent person (rejecting the null hypothesis that the person is innocent when they are), it is a Type 1 error.

**Type 2 Error Example:**
Conversely, if you fail to convict a guilty person (failing to reject the null hypothesis that the person is innocent when they are not), it is a Type 2 error.

**Sample Problem:**

### Q7: 95% Confidence Interval Calculation

Given a sample mean of 50, standard deviation of 5, and a sample size of 100, you can use the formula:

\[ \text{Confidence Interval} = \text{Sample Mean} \pm \left(\text{Critical Value} \times \frac{\text{Standard Deviation}}{\sqrt{\text{Sample Size}}}\right) \]

Interpretation: We are 95% confident that the true population mean falls within this interval.

### Q9: Z-Score Calculation and Interpretation

\[ Z = \frac{X - \mu}{\sigma} \]

Given a data point value of 75, a population mean of 70, and a population standard deviation of 5:

\[ Z = \frac{75 - 70}{5} = 1 \]

Interpretation: The data point is 1 standard deviation above the mean.
