## Q1

Scatter Plot (sns.scatterplot):
Use: Display the relationship between two continuous variables.
Line Plot (sns.lineplot):
Use: Show the trend of a continuous variable over a continuous interval or time.
Bar Plot (sns.barplot):
Use: Compare the values of different categories or groups.
Box Plot (sns.boxplot):
Use: Visualize the distribution of data and identify outliers in a dataset.
Pair Plot (sns.pairplot):
Use: Display pairwise relationships in a dataset, useful for exploring correlations and distributions simultaneously.

```python
## Q2
import seaborn as sns
import matplotlib.pyplot as plt

# Load the fmri dataset
fmri_data = sns.load_dataset("fmri")

# Line plot
sns.lineplot(x="timepoint", y="signal",  hue="event", style="region",data=fmri_data)
plt.title('Line Plot of fmri Data')
plt.show()
```

```python
## Q3
# Load the titanic dataset
titanic_data = sns.load_dataset("titanic")

# Box plots
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.boxplot(x="pclass", y="age", data=titanic_data)
plt.title('Box Plot of Age by Pclass')

plt.subplot(1, 2, 2)
sns.boxplot(x="pclass", y="fare", data=titanic_data)
plt.title('Box Plot of Fare by Pclass')

plt.show()
```

```python
## Q4
# Load the diamonds dataset
diamonds_data = sns.load_dataset("diamonds")

# Histogram with hue
sns.histplot(data=diamonds_data, x="price", hue="cut", bins=30, kde=True)
plt.title('Histogram of Price by Cut')
plt.show()
```

```python
## Q5
# Load the iris dataset
iris_data = sns.load_dataset("iris")

# Pair plot with hue
sns.pairplot(iris_data, hue="species")
plt.title('Pair Plot of Iris Dataset')
plt.show()
```

```python
## Q6
# Load the flights dataset
flights_data = sns.load_dataset("flights")

# Pivot the data to create a matrix for the heatmap
flights_pivot = flights_data.pivot_table(index='month', columns='year', values='passengers')

# Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(flights_pivot, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Flights Dataset - Heatmap')
plt.show()
```
